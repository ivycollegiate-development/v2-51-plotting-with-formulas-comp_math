```python
import matplotlib.pyplot as plt # Imports the pyplot function in the matplotlib library and names it plt

# A couple of basic tuples or immutable lists
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
# Creates a basic plot
plt.plot(x_numbers, y_numbers)
# Draw an x-axis label
plt.xlabel('Time (seconds)')
# Draw a y-axis label
plt.ylabel('Distance (meters)')
# Add a Title
plt.title('My Distance-Time Graph')
# Plots the actual graph
plt.show()
```
