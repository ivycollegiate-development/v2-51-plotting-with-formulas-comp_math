```python
import matplotlib.pyplot as plt

x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]

plt.plot(x_numbers, y_numbers)

plt.xlabel('Time(second)')
plt.ylabel('Distance(meters)')
plt.title('My Distance-Time Graph')

plt.show()

```
